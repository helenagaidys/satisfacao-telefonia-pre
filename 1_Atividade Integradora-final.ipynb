#### Banco de Dados da Pesquisa - Telefonia Móvel (Pré-paga)

A Agência Nacional de Telecomunicações (Anatel) foi criada pela Lei 9.472, de 16 de julho de 1997 – mais conhecida como Lei Geral de Telecomunicações (LGT), sendo a primeira agência reguladora a ser instalada no Brasil, em 5 de novembro daquele mesmo ano. 

Segundo a [Anatel](https://www.anatel.gov.br/consumidor/pesquisa-de-satisfacao-e-qualidade), "desde 2015, a Anatel realiza anualmente pesquisas que medem simultaneamente a satisfação e a qualidade percebida pelos consumidores dos serviços de telecomunicações para os serviços de telefonia fixa, telefonia móvel (pré e pós), banda larga fixa e TV por assinatura".

Esse banco de dados refere-se à pesquisa de usuários de telefones celulares pré pagos, nos anos de 2015 a 2019.

##### Descrição da Pesquisa

Desenho da amostra: A seleção integralmente aleatória das amostras foi realizada pela Anatel a partir de sorteio da base de dados de Códigos de Acesso em operação fornecida pelas prestadoras, dentro dos segmentos: prestadora, UF do consumidor e modalidade do
serviço.

O questionário foi aplicado por telefone (CATI). "*CATI (Computer Assisted Telephone Interviewing) é a técnica de pesquisa realizada através do telefone. O entrevistador é orientado durante a entrevista seguindo um algoritmo disponibilizado pelo computador. Esse software tem a capacidade de configurar a ordem e o tipo de questões baseando-se nas respostas anteriores, ou na informação conhecida sobre o participante"* (https://pt.wikipedia.org/wiki/CATI)

Segundo os relatórios da [Anatel](https://www.anatel.gov.br/), foram realizadas 43.970 entrevistas em [2015](https://www.anatel.gov.br/institucional/index.php?option=com_anexarlink&hash=8782d9096582094f259915f8e947bfd4), 29.802 em [2016](https://www.anatel.gov.br/institucional/index.php?option=com_anexarlink&hash=c49408ae106fb11a9a4e2213e4b00dd7), 30.666 em [2017](https://www.anatel.gov.br/institucional/index.php?option=com_anexarlink&hash=c49408ae106fb11a9a4e2213e4b00dd7), 23.973 em [2018](https://www.anatel.gov.br/institucional/index.php?option=com_anexarlink&hash=ab0927bb7cb98d0cac887181ae16f91a) e 19.609 entrevistas em [2019](https://www.anatel.gov.br/institucional/index.php?option=com_anexarlink&hash=646b95696dc1926c6b94bb05e849d1db).

### Satisfação geral dos entrevistados

Conforme o enunciado da atividade, consideramos satisfação "alta" notas iguais ou maiores que 8 e satisfação "baixa" notas iguais ou menores que 7.

```python
# Verificando se J1 tem algum resultado fora do esperado

def count_elements(seq) -> dict:

    hist = {}
    for i in seq:
        hist[i] = hist.get(i, 0) + 1
    return hist

satisfacoes = count_elements(df["J1"])
satisfacoes
```

```python
names = list(satisfacoes.keys())
values = list(satisfacoes.values())

fig, axs = plt.subplots(figsize=(9, 3))
axs.bar(names, values)
fig.suptitle('Quantidade de participantes por nota de satisfação geral')
```

```python
# Criando função para separar o título grande em um bloco com quebra de linhas

def split_title_line(title_text, split_on='(', max_words=5):  # , max_words=None):
    """
    A function that splits any string based on specific character
    (returning it with the string), with maximum number of words on it
    """
    split_at = title_text.find (split_on)
    ti = title_text
    if split_at > 1:
        ti = ti.split (split_on)
        for i, tx in enumerate (ti[1:]):
            ti[i + 1] = split_on + tx
    if type (ti) == type ('text'):
        ti = [ti]
    for j, td in enumerate (ti):
        if td.find (split_on) > 0:
            pass
        else:
            tw = td.split ()
            t2 = []
            for i in range (0, len (tw), max_words):
                t2.append (' '.join (tw[i:max_words + i]))
            ti[j] = t2
    ti = [item for sublist in ti for item in sublist]
    ret_tex = []
    for j in range (len (ti)):
        for i in range(0, len(ti)-1, 2):
            if len (ti[i].split()) + len (ti[i+1].split ()) <= max_words:
                mrg = " ".join ([ti[i], ti[i+1]])
                ti = [mrg] + ti[2:]
                break

    if len (ti[-2].split ()) + len (ti[-1].split ()) <= max_words:
        mrg = " ".join ([ti[-2], ti[-1]])
        ti = ti[:-2] + [mrg]
    return '\n'.join (ti)
```

Conforme [dados da Anatel](https://www.anatel.gov.br/paineis/acessos), hoje há 226 milhões de telefones móveis no Brasil. Atualmente, os celulares pré pagos respondem por 50,8% e 49,2% são pós pagos. As projeções são que ainda neste ano de 2020, os celulares pós-pagos devam superar os planos pré-pagos. 

Observa-se que esta mudança de comportamento do consumidor vem ocorrendo desde 2015 (ano da primeira pesquisa). Em [2015](https://www.anatel.gov.br/paineis/acessos/telefonia-movel), haviam 284 milhões de linhas de celular, sendo que 75% eram pré-pagos. Ao longo dos anos, houve diminuição do número de linhas. Em 2015, era frequente que um usuário tivesse mais de uma linha de celular (chip) de diferentes operadoras, pois havia diferença de tarifa para as chamadas de voz e de mensagens de texto (SMS) de operadora para outra. Com a evolução tecnológica e o uso de aplicativos de mensagens e que fazem ligação de voz e vídeo, os usuários passaram a optar por manter somente uma linha de celular. A migração de planos pré-pagos para pós-pagos tem sido crescente desde 2015. O uso cada vez maior da internet no celular e a oferta de planos mais vantajosos explicam essa mudança de padrão.

As pesquisas nos anos de 2015 a 2019 já apontavam que o grau de satisfação do usuário tinha correlação direta com a facilidade de entendimento dos planos e serviços contratados.

A modelagem será feita em R para identificar quais as variáveis que têm maior correlação e um modelo preditivo da satisfação do usuário.
